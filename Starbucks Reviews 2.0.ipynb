### Distribution of Ages, Occupation and Annual Income

```python
# array of useful columns needed for this analysis
customerDetails = ['2. Your Age', '3. Are you currently....?', '4. What is your annual income?']
Genders = ['1. Your Gender']

# declaring the subplot
fig, sp = plt.subplots(len(customerDetails), 2, figsize=(14, 10))

for i, col in enumerate(customerDetails):
    # Get value counts for Male and Female
    male_counts = data[data['1. Your Gender'] == 'Male'][col].value_counts()
    female_counts = data[data['1. Your Gender'] == 'Female'][col].value_counts()

    # Plot a pie chart on the 1st column
    sp[i, 0].pie(male_counts, labels=male_counts.index, autopct='%1.1f%%', startangle=45)
    sp[i, 0].set_title(f'Male - {col}')
    sp[i, 0].axis('equal')

    # Plot a pie chart on the 2nd column
    sp[i, 1].pie(female_counts, labels=female_counts.index, autopct='%1.1f%%', startangle=45)
    sp[i, 1].set_title(f'Female - {col}')
    sp[i, 1].axis('equal')

# Adjust the layout
plt.tight_layout()
plt.show()
```

### Customers rating their Starbucks experience in stores

```python
experienceData = ['13. How would you rate the price range at Starbucks?', '14. How important are sales and promotions in your purchase decision?', 
                  '15. How would you rate the ambiance at Starbucks? (lighting, music, etc...)', '16. You rate the WiFi quality at Starbucks as..',
                  '17. How would you rate the service at Starbucks? (Promptness, friendliness, etc..)',]

for col in experienceData:
    plt.figure(figsize=(12, 6))
    
    sbn.histplot(data = data, x = col, bins=range(int(data[col].min()), int(data[col].max()) + 1), kde = True)
    plt.title(col)
    plt.xlabel("Ratings")
    
    plt.grid()
    plt.show()
```

### Starbucks Membership

```python
chanceData = data['9. Do you have Starbucks membership card?']
color = ["Green", "Red"]

plt.figure(figsize=(12, 6))

sbn.countplot(data = data, x = '1. Your Gender', hue = chanceData , palette = color)

plt.title("Starbucks Membership")
plt.legend()
plt.show()
```

### Those with memberships, who are most likely to come back?

```python
filteredDataYes = data[(data['9. Do you have Starbucks membership card?'] == 'Yes')]
filteredDataNo = data[(data['9. Do you have Starbucks membership card?'] == 'No')]

comebackData = data['20. Will you continue buying at Starbucks?']

plt.figure(figsize=(12, 6))

#calls a count plot that uses the filtered data. The x is using the gender column and the hue is focusing on the last column of the dataset.
sbn.countplot(data = filteredDataYes, x = '1. Your Gender', hue = comebackData, palette = color)

plt.title("Membership users who would come back")
plt.legend()
plt.show()
```

### Those without memberships, who are most likely to come back?

```python
plt.figure(figsize=(12, 6))

sbn.countplot(data = filteredDataNo, x = '1. Your Gender', hue = comebackData, palette = color)

plt.title("Non Membership users who would come back")
plt.legend()
plt.show()
```
